# This part is the machine learning aspect of our project.

I will be using corelation matrix to determine how diffferent feachures are related to each other.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib

# Identify categorical and numeric columns
categorical_columns = ['Sex', 'Diet', 'Country', 'Continent', 'Hemisphere', 'Heart Attack Risk', 'Blood Pressure']
numeric_columns = ['Age', 'Cholesterol', 'Heart Rate', 'Income', 'BMI', 'Triglycerides', 'Exercise Hours Per Week', 'Sedentary Hours Per Day']

# Encode categorical columns using LabelEncoder
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    dfheart[col] = le.fit_transform(dfheart[col])
    label_encoders[col] = le

# Define a ColumnTransformer with parallel processing
column_transformer = ColumnTransformer(
    transformers=[
        ('categorical', 'passthrough', categorical_columns),
        ('numeric', StandardScaler(), numeric_columns)
    ],
    n_jobs=-1  # Use all available CPU cores
)

# Create a pipeline to apply the transformations
preprocessing_pipeline = Pipeline([
    ('transform', column_transformer)
])

# Fit and transform the data in parallel
preprocessed_data = preprocessing_pipeline.fit_transform(dfheart)

# Convert the result back to a DataFrame with column names
preprocessed_data = pd.DataFrame(preprocessed_data, columns=categorical_columns + numeric_columns)


```

```python
# Calculate the correlation matrix
correlation_matrix = preprocessed_data.corr()

# Display the correlation matrix
print(correlation_matrix)
```

This part of machine learinign is forming a neural netorwrk to predic Cholecap prices for the next 10 weeks

```python


dfweekly= pd.read_csv("/Users/Karim/Documents/VeevaChallenge/data/salesweekly.csv")

# Select features and target variable
features = dfweekly[['Simvastatin', 'Fluvastatin', 'Cerivastatin', 'Rosuvastatin calcium', 'Pravastatin', 'Pannorin', 'Pitavastatin']]
target = dfweekly['Cholecap']  # You can choose a different target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build a simple feedforward neural network model
model = keras.Sequential([
    layers.Input(shape=(7,)),  # Adjust input shape according to your feature count
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),    
    layers.Dense(1)  # Output layer for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
from sklearn.metrics import r2_score
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_log_error
msle = mean_squared_log_error(y_test, y_pred)
print("Mean Squared Logarithmic Error:", msle)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
model.save("kerasmodel.h5")

```

```python
model = keras.models.load_model('kerasmodel.h5')
input_data = dfweekly.iloc[-10:, 1:8]
input_data = scaler.transform(input_data)
predicted_values = model.predict(input_data)
date_range = pd.date_range(start=dfweekly['datum'].iloc[-1], periods=10, freq='W')
if len(date_range) == len(predicted_values):
    # Create a DataFrame with the predicted values and dates
    predicted_df = pd.DataFrame({'Date': date_range, 'Cholecap (Predicted)': predicted_values.flatten()})
    predicted_df['negative Difference with MSE'] = predicted_df['Cholecap (Predicted)'] - mse
    predicted_df['positive Difference with MSE'] = predicted_df['Cholecap (Predicted)'] + mse
    # Plot the predicted data using Plotly
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=date_range, y=predicted_df['Cholecap (Predicted)'], mode='lines', name='Predicted price'))
    fig.add_trace(go.Scatter(x=date_range, y=predicted_df['negative Difference with MSE'], mode='lines', marker=dict(color='red'),name='Negative MSE'))
    fig.add_trace(go.Scatter(x=date_range, y=predicted_df['positive Difference with MSE'], mode='lines', marker=dict(color='red'),fill='tonexty', name='Positive MSE'))
    fig.update_layout(title='Predicted Cholecap Sales for Next 10 Weeks')
    fig.update_xaxes(title_text='Date')  # X-axis label
    fig.update_yaxes(title_text='Cholecap Sales ($1000)')  # Y-axis label
    fig.show()
else:
    print("Date range and predicted values have different lengths. Check your data.")

print(len(predicted_values))
```
